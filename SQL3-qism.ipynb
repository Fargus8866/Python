```python
pd.read_sql("SELECT * FROM elect LIMIT 15",db,index_col='index')
```

```python
query="SELECT DISTINCT country FROM elect "
countries=pd.read_sql(query,db)
```

```python
'Uzbekistan' in countries.country.values
```

```python
query="SELECT * FROM elect WHERE country='Uzbekistan'"
pd.read_sql(query,db).head()
```

```python
query="SELECT * FROM elect WHERE country='Uzbekistan' AND year>2010"
pd.read_sql(query,db)
```

```python
query="SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country='Uzbekistan' AND year>2010"
pd.read_sql(query,db)
```

```python
query = "SELECT * FROM elect WHERE country='Uzbekistan' AND year BETWEEN 2010 AND 2021"
pd.read_sql(query, db)
```

```python
query = "SELECT country,year,coal,gas,hydro,oil FROM elect WHERE country='Uzbekistan' AND year IN (1990,1995,2000,2005,2010)"
pd.read_sql(query, db)
```

```python
query = "SELECT country,year,coal,gas,hydro,oil FROM elect WHERE code LIKE 'U__' AND year=2020"
pd.read_sql(query, db)
```

```python
query = "SELECT * FROM elect WHERE code LIKE 'U__' "
pd.read_sql(query, db)
```
